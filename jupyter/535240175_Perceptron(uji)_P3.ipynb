```python
import numpy as np 

x = np.array([
    [1, 1],
    [1, -1],
    [-1, 1],
    [-1, -1],
])

t = np.array([1, 1, 1, -1])

w1, w2, b = 0, 0, 0
alpha = 1
threshold = 0
epoch = 1

def activation(y_in):
    return 1 if y_in > threshold else (-1 if y_in < threshold else 0)

print("Epoch\tx1\tx2\ty_in\ty\tt\t△w1\t△w2\t△b\tw1\tw2\tb")
while True:
    error = False
    for i in range(len(x)):
        x1, x2 = x[i]
        y_in = b + (w1 * x1) + (w2 * x2)
        y = activation(y_in)

        delta_w1, delta_w2, delta_b = 0, 0, 0

        if y != t[i]:
            delta_w1 = alpha * t[i] * x1
            delta_w2 = alpha * t[i]  * x2
            delta_b = alpha * t[i]
            
            w1 += delta_w1
            w2 += delta_w2
            b += delta_b

            error = True
        
        print(f"{epoch}\t{x1}\t{x2}\t{y_in}\
                t{y}\t{t[i]}\t{delta_w1}\t{delta_w2}\t{delta_b}\t{w1}\t{w2}\t{b}")
    if not error:
        break

    epoch += 1

print(f"\nKonvergen pada epoch {epoch}")
print(f"Bobot akhir: w1={w1}, w2={w2}, b={b}")
```

```python
print("=== pengujian ===")
x1 = int(input("Masukkan nilai x1: "))
x2 = int(input("Masukkan nilai x2: "))


print(f"Masukkan x1 (1 atau 0): {x1}")
print(f"Masukkan x2 (1 atau 0): {x2}")

y_in = b + (w1 * x1) + (w2 * x2)
y_prediksi = activation(y_in)

print(f"\nHasil:")
print(f"y_in: {y_in}")
print(f"Output (y): {y_prediksi}")
```

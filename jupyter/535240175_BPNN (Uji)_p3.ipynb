```python
import math

def sigmoid(x):
    return 1 / (1 + math.exp(-x))
```

```python

x1_train, x2_train = 1, 0
t = 0
alpha = 1

v0, v1, v2 = 0.1, 0.1, 0.4
w0, w1 = 0.5, 1

z_in = v0 + (v1 * x1_train) + (v2 * x2_train)
z = sigmoid(z_in)

y_in_train = w0 + (w1 * z)
y_train = sigmoid(y_in_train)

dk = (t - y_train) * y_train * (1 - y_train)
dw1 = alpha * dk * z
dw0 = alpha * dk * 1

d_in_j = dk * w1
dj = d_in_j * (z * (1 - z))
dv1 = alpha * dj * x1_train
dv2 = alpha * dj * x2_train
dv0 = alpha * dj * 1

w1_baru = w1 + dw1
w0_baru = w0 + dw0
v1_baru = v1 + dv1
v2_baru = v2 + dv2
v0_baru = v0 + dv0

print("=== PROSES TRAINING SELESAI ===")
print(f"Bobot W baru: {w0_baru:.4f}, {w1_baru:.4f}")
print(f"Bobot V baru: {v0_baru:.4f}, {v1_baru:.4f}, {v2_baru:.4f}")
print("-" * 30)
```

```python

print("=== PENGUJIAN MODEL ===")
x1_test = int(input("Masukkan nilai x1 (1 atau 0): "))
x2_test = int(input("Masukkan nilai x2 (1 atau 0): "))

print(f"Masukkan x1 (1 atau 0): {x1_test}")
print(f"Masukkan x2 (1 atau 0): {x2_test}")
# Forward pass dengan bobot baru dan input baru
z_in_test = v0_baru + (v1_baru * x1_test) + (v2_baru * x2_test)
z_test = sigmoid(z_in_test)

y_in_test = w0_baru + (w1_baru * z_test)
y_final = sigmoid(y_in_test)

# Menentukan Prediksi Kelas (Threshold 0.5)
y_prediksi = 1 if y_final >= 0.5 else 0

print("\nHasil:")
print(f"y (Sigmoid): {y_final:.3f}")
print(f"output_akhir (kelas): {y_prediksi}")
```
